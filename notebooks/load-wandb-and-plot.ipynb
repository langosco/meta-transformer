```python
from pathlib import Path
import pprint
import matplotlib.pyplot as plt
from dataclasses import asdict
import numpy as np
import wandb
import seaborn as sns
sns.set()
```

```python
# load data from wandb 
api = wandb.Api()
runs = api.runs("lauro/detect-backdoors-cifar10")
run = runs[0]
```

```python
for r in runs:
    if r.name == "breezy-lion-90":
        run = r
```

```python
plt.figure(figsize=[7, 3])
plt.bar(["Depoisoned (iid)", "Depoisoned (OOD) [TODO]", "Depoisoned (OOD 2) [TODO]"], 
        [run.summary["val/accuracy"],  0.5, 0.5],
        color=["tab:blue", "tab:gray", "tab:gray"])
plt.tight_layout()
plt.savefig("backdoor-detection-accuracy.pdf")
```

```python
# Data
x = ['A', 'B', 'C', 'D']
y = [3, 7, 1, 10]

# Create basic bar chart
plt.bar(x, y, color='blue')

# 1. Add labels and titles
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Example Bar Chart')

# 2. Change Bar Colors
colors = ['red', 'green', 'blue', 'yellow']
plt.bar(x, y, color=colors)

# 3. Add Grid Lines
plt.grid(True, axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# 4. Add Data Labels
for i, value in enumerate(y):
    plt.text(i, value + 0.1, str(value), ha='center')

# 9. Legends
plt.legend(['Dataset 1'])

# 10. Styles
plt.style.use('seaborn')

plt.show()
```
